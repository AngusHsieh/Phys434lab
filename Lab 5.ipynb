## 5a

```python
infile1 = open ("qcd_100000_pt_1000_1200.pkl", 'rb')
qcd_dict = pickle.load(infile1)
print(qcd_dict.keys())

infile2 = open ("higgs_100000_pt_1000_1200.pkl", 'rb')
higgs_dict = pickle.load(infile2)
print(higgs_dict.keys())

qcd = np.array(qcd_dict)
higgs = np.array(higgs_dict)
```

```python
for i in range(10):
    print(qcd_dict['mass'][i], qcd_dict['d2'][i])
```

## 5b.1

```python
for x in range(0, 14):
    fig, ax = plt.subplots()
    title = higgs_dict.keys()[x]
    ax.set_title("Histogram of " + str(title))
    ax.hist(qcd[:,x], bins = 100, density = True, alpha = 0.7)
    ax.hist(higgs[:,x], bins = 100, density = True, alpha = 0.7)
    ax.legend(['Background', 'Signal'])
    ax.set_xlabel('Measurements')
    ax.set_ylabel('Probability')
    plt.show()
```

Not all features provide discrimination power between signal(higgs) and background (qcd). Phi and t32 don't provide discrimination power.

```python
ht = pd.DataFrame(higgs_dict)
ht.corr()
```

```python
binsize = 500
plt.hist2d(higgs_dict['t2'], higgs_dict['eta'], bins = [binsize, binsize], cmap=plt.cm.jet)
plt.title('Higgs eta vs t2 Correlation')
plt.xlabel('t2')
plt.ylabel('eta')
plt.colorbar()
plt.figure()
plt.hist2d(higgs_dict['t2'], higgs_dict['t3'], bins = [binsize, binsize], cmap=plt.cm.jet)
plt.title('Higgs t3 vs t2 Correlation')
plt.xlabel('t2')
plt.ylabel('t3')
plt.colorbar()
plt.figure()
plt.hist2d(higgs_dict['t2'], higgs_dict['t21'], bins = [binsize, binsize], cmap=plt.cm.jet)
plt.title('Higgs t21 vs t2 Correlation')
plt.xlabel('t2')
plt.ylabel('t21')
plt.colorbar()
plt.figure()
plt.hist2d(higgs_dict['t3'], higgs_dict['t21'], bins = [binsize, binsize], cmap=plt.cm.jet)
plt.title('Higgs t21 vs t3 Correlation')
plt.xlabel('t3')
plt.ylabel('t21')
plt.colorbar()
plt.figure()
```

First graph is a randomly picked variables graph and it is pretty scattered. On the other hands, the next three graphs clearly have the linear correlation.

```python
qt = pd.DataFrame(qcd_dict)
qt.corr()
```

qcd has the same correlation as higgs.

```python
binsize = 500
plt.hist2d(qcd_dict['t2'], qcd_dict['eta'], bins = [binsize, binsize], cmap=plt.cm.jet)
plt.title('Qcd eta vs t2 Correlation')
plt.xlabel('t2')
plt.ylabel('eta')
plt.colorbar()
plt.figure()
plt.hist2d(qcd_dict['t2'], qcd_dict['t3'], bins = [binsize, binsize], cmap=plt.cm.jet)
plt.title('Qcd t3 vs t2 Correlation')
plt.xlabel('t2')
plt.ylabel('t3')
plt.colorbar()
plt.figure()
plt.hist2d(qcd_dict['t2'], qcd_dict['t21'], bins = [binsize, binsize], cmap=plt.cm.jet)
plt.title('Qcd t21 vs t2 Correlation')
plt.xlabel('t2')
plt.ylabel('t21')
plt.colorbar()
plt.figure()
plt.hist2d(qcd_dict['t3'], qcd_dict['t21'], bins = [binsize, binsize], cmap=plt.cm.jet)
plt.title('Qcd t21 vs t3 Correlation')
plt.xlabel('t3')
plt.ylabel('t21')
plt.colorbar()
plt.figure()
```

As shown above, the 3 graphs all have a stacked line, which suggests a correlation.
